```python
summerlist = []
#start tripdata05-June, stop tripdata07-August
for i in range(5,8):
    df=pd.read_csv(f"nytripdata/tripdata0{i}.csv")
    summerlist.append(df)
    
summerdf = pd.concat(summerlist)
summerdf
```

```python
winterlist = []
winter = [0,1,11]
#include tripdata00-January,tripdata01-February,tripdata11-December
for i in winter:
    df=pd.read_csv(f"nytripdata/tripdata0{i}.csv")
    winterlist.append(df)
    
winterdf = pd.concat(winterlist)
winterdf
```

```python
#calcuate the age by subtracting current year 2020 and create new column "age" 
summerdf["age"] = 2020-summerdf["birth year"]
winterdf["age"] = 2020-winterdf["birth year"]
```

```python
#replace 0,1,2 in summerdf with unknown,M,F
summerdf.loc[summerdf["gender"]==0,"gender"]="unknown"
summerdf.loc[summerdf["gender"]==1,"gender"]="M"
summerdf.loc[summerdf["gender"]==2,"gender"]="F"
```

```python
#replace 0,1,2 in winterdf with unknown,M,F
winterdf.loc[winterdf["gender"]==0,"gender"]="unknown"
winterdf.loc[winterdf["gender"]==1,"gender"]="M"
winterdf.loc[winterdf["gender"]==2,"gender"]="F"
```

```python
#calcuate the minutes and create new column "minutes" 
summerdf["minutes"] = summerdf["tripduration"]/60
winterdf["minutes"] = winterdf["tripduration"]/60
```

```python
summerdf.to_csv("summer.csv", index=False)
```
